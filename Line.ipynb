```python
# Creating EGA
# per-pack difference (actual - target)
df['ega_per_pack'] = df['pack_grammage'] - df['target_weight']


# total giveaway per record (per-pack * count)df
df['ega_total_gm'] = df['ega_per_pack'] * df['actual_count']


# percent giveaway relative to target (per pack)
df['ega_total_per'] = (df['ega_per_pack'] / df['target_weight']) * 100

df
```

EGA (Extra Give Away) per Pack

Target Weight â†’ The intended standard weight of product per pack (e.g., 100g).

EGA per Pack = Actual Weight â€“ Target Weight

If EGA > 0 â†’ Overfill (extra product given away per pack).

If EGA < 0 â†’ Underfill (pack weight less than target, often non-compliant).

If EGA = 0 â†’ Perfect fill (no giveaway).

EGA Total (Batch/Interval)

Formula:
EGA Total = EGA per Pack Ã— Number of Packs in the Batch

This represents the absolute grams of product given away (or lost) in that batch or time interval.

Useful for:

Cost analysis â†’ How much product is lost as free giveaway.

Efficiency check â†’ Detecting whether certain machines, shifts, or pack sizes cause higher giveaway.

Pattern analysis â†’ Compare across pack sizes, time intervals, or production lines.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['actual_weight'].dropna().reset_index(drop=True),
         color='red', linewidth=1.5, label='Actual Weight')

plt.title('Actual Pack Weight Line Plot')
plt.xlabel('Pack Number (sequence)')
plt.ylabel('Weight in gm')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

**Description of the Actual Pack Weight Line Plot**

This line plot visualizes the 'Actual Weight' of each pack in the dataset over its sequence number. Each point on the line represents the measured weight of an individual pack, plotted against the order in which it appears in the filtered data.

*   **X-axis:** Represents the "Pack Number (sequence)", indicating the order of the packs as recorded in the dataset after handling any missing values. It does not necessarily correspond to a specific time interval but rather the sequence of measurements.
*   **Y-axis:** Represents the "Weight in gm", showing the actual measured weight of each individual pack in grams.
*   **The Red Line:** Connects the actual weight measurements of consecutive packs.

Analyzing this plot can help to:

*   Identify trends or patterns in pack weight over time or sequence.
*   Spot significant deviations from the target weight (although the target weight is not explicitly shown on this plot).
*   Observe the variability or consistency of the filling process.
*   Detect potential issues like gradual drift in weight, sudden drops or spikes, or cyclical variations.

Comparing this plot to the target weight (if overlaid) and analyzing the EGA (Extra Give Away) metrics calculated earlier can provide a comprehensive understanding of the weighing and packaging process performance.

```python
# over fill
positive_ega = df[df['ega_total_gm'] > 0]
positive_count = positive_ega.shape[0]

# underfill
negative_ega = df[df['ega_total_gm'] < 0]
negative_count = negative_ega.shape[0]

# Zero EGA (exact match=actual)
zero_ega = df[df['ega_total_gm'] == 0]
zero_count = zero_ega.shape[0]

# Print results
print(f"Number of overfilled records (ega_total_gm > 0): {positive_count}")
print(f"Number of underfilled records (ega_total_gm < 0): {negative_count}")
print(f"Number of exact match records (ega_total_gm = 0): {zero_count}")
```

```python
import matplotlib.pyplot as plt

categories = ['Overfill', 'Underfill', 'Exact Match']
counts = [positive_count, negative_count, zero_count]

plt.figure(figsize=(8,5))
plt.plot(categories, counts, marker='o', linestyle='-', linewidth=2, markersize=8,
         color='purple', label='EGA Distribution')

plt.title('EGA Total Giveaway Distribution', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Records', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
